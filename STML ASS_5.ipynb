```python
for column in features.columns:
    plt.figure(figsize=(10, 6))
    plt.scatter(features[column], target_var, alpha=0.5)
    plt.xlabel(column)
    plt.ylabel('Target')
    plt.title(f'Relationship between {column} and Target')
    plt.show()
```

```python

one_values = ["Male", "Yes"]
zero_values = ["Female","No"]

for column in features.columns:
    features[column] = features[column].replace(to_replace=one_values, value=1)
    features[column] = features[column].replace(to_replace=zero_values, value=0)
```
